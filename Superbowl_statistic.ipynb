4.	What is the trend for tweet volume?

```python
#!/usr/bin/python
# -*- coding: utf-8 -*-
import MySQLdb as mdb
import pandas as pd
import sys
def trend_2014():
    try:
        con = mdb.connect('128.206.116.195', 'tg4_ro', '?3stEt7!3hUbRa-R', 'tw4_db');

        cur = con.cursor()
        #cur.execute("SELECT created_at, from_user, from_user_followers, text from tweet where job_id=1196")
        cur.execute("Select month(created_at) as theMonth, count(*) as counter from tweet where job_id=187 and year(created_at) = 2014 group by theMonth"),

        ver = cur.fetchall()
        list_tuple = []
        for row in ver:
            list_tuple.append(row)
        return pd.DataFrame(list_tuple, columns = ['month','counter'])

    except mdb.Error as e:

        print ("Error %d: %s" % (e.args[0],e.args[1]))
        sys.exit(1)

    finally:

        if con:
            con.close()
            
df_2014 = trend_2014()
print(df_2014)
```

```python
df_2014.to_csv('trend2014.csv', sep='\t', encoding='utf-8')
```

```python
#!/usr/bin/python
# -*- coding: utf-8 -*-
import MySQLdb as mdb
import pandas as pd
import sys
def Beyonce():
    try:
        con = mdb.connect('128.206.116.195', 'tg4_ro', '?3stEt7!3hUbRa-R', 'tw4_db');

        cur = con.cursor()
        #cur.execute("SELECT created_at, from_user, from_user_followers, text from tweet where job_id=1196")
        cur.execute("Select hour(created_at) as theHour, count(*) as counter from tweet where job_id=187 and text like '%Beyonce%'and year(created_at) = 2016 and month(created_at) = 2 and day(created_at) = 7 group by theHour"),

        ver = cur.fetchall()
        list_tuple = []
        for row in ver:
            list_tuple.append(row)
        return pd.DataFrame(list_tuple, columns = ['hour','counter'])

    except mdb.Error as e:

        print ("Error %d: %s" % (e.args[0],e.args[1]))
        sys.exit(1)

    finally:

        if con:
            con.close()
            
df_beyonce = Beyonce()
print(df_beyonce)
```

1.	How frequently is each user mentioned during each month of the lifecycle?

```python
#!/usr/bin/python
# -*- coding: utf-8 -*-
import MySQLdb as mdb
import pandas as pd
import sys
def frequency():
    try:
        con = mdb.connect('128.206.116.195', 'tg4_ro', '?3stEt7!3hUbRa-R', 'tw4_db');

        cur = con.cursor()
        #cur.execute("SELECT created_at, from_user, from_user_followers, text from tweet where job_id=1196")
        cur.execute("Select month(created_at), year(created_at), count(*) from tweet where job_id = 187 and to_user is not null group by year(created_at) , month(created_at)"),

        ver = cur.fetchall()
        list_tuple = []
        for row in ver:
            list_tuple.append(row)
        return pd.DataFrame(list_tuple, columns = ['month','year','counter'])

    except mdb.Error as e:

        print ("Error %d: %s" % (e.args[0],e.args[1]))
        sys.exit(1)

    finally:

        if con:
            con.close()
            
df_frequency = frequency()
print(df_frequency)
```
