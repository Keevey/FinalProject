```python
#!/usr/bin/python
# -*- coding: utf-8 -*-
import MySQLdb as mdb
import pandas as pd
import sys
import operator
import re
def trend_2014():
    try:
        con = mdb.connect('128.206.116.195', 'tg4_ro', '?3stEt7!3hUbRa-R', 'tw4_db');

        cur = con.cursor()
        #cur.execute("SELECT created_at, from_user, from_user_followers, text from tweet where job_id=1196")
        cur.execute("Select from_user, from_user_followers, from_user_following, from_user_favorites, from_user_tweets,text from tweet where job_id=187 limit 10000000"),

        
        halftime = ['beyonce','rihanna','breaking','halftime','advertising','katyperry','kenny','phoenix','bruno']
        game = ['aaronrodgers','packers','curvedview','greenbay','seahawks','patriots','broncos','steelers','panther','gohawks','denver','seattle']
        halftime_cnt = 0
        game_cnt = 0
        list_tuple = []
        
        row = cur.fetchone()
        while row is not None:
            #for item in row:
                #list_tuple.append(item)
            for item in row:
                if type(item) == int:
                    continue
                for i in item.split():
                    if i.startswith("#"):
                        i = re.sub(r'[^a-zA-Z0-9\s]+','',i).lower().strip()
                        if i in halftime:
                            halftime_cnt = halftime_cnt + 1
                        elif i in game:
                            game_cnt = game_cnt + 1
            if halftime_cnt > 0 and game_cnt == 0: 
                #list_tuple.append("only halftime")
                row = row + ('only halftime',)
            elif halftime_cnt == 0 and game_cnt > 0:
                #list_tuple.append("only game")
                row = row + ('only game',)
            elif halftime_cnt > game_cnt:
                #list_tuple.append("both but more halftime")
                row = row + ('both but more halftime',)
            elif halftime_cnt < game_cnt:
                #list_tuple.append("both but more game")
                row = row + ('both but more game',)
            elif halftime_cnt == game_cnt and halftime_cnt != 0 and game_cnt != 0:
                #list_tuple.append("both")
                row = row + ('both',)
            else:
                #list_tuple.append("no hashtags")
                row = row + ('no hashtags',)
            list_tuple.append(row)
            row = cur.fetchone()
            
            
        return pd.DataFrame(list_tuple, columns = ['user','followers','following','favorites','total_tweets','text','class_people'])

    except mdb.Error as e:

        print ("Error %d: %s" % (e.args[0],e.args[1]))
        sys.exit(1)

    finally:

        if con:
            con.close()
            
df_2014 = trend_2014()
print(df_2014)
```
